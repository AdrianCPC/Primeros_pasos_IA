```python
def eliminar(textos, regex):
  if type(textos) == str:
    return regex.sub('', textos)
  else:
    return [regex.sub('', texto) for texto in textos]
```

```python
regex_html = re.compile(r'<.*?>')
pregunta_sin_tag = eliminar(pregunta_portugues, regex_html)
print(pregunta_sin_tag)
```

```python
regex_html = re.compile(r'<.*?>')
pregunta_sin_tag = eliminar(pregunta_español, regex_html)
print(pregunta_sin_tag)
```

```python
regex_html = re.compile(r'<.*?>')
pregunta_sin_tag = eliminar(pregunta_ingles, regex_html)
print(pregunta_sin_tag)
```

```python
print(pregunta_ingles)
```

```python
def sustituir_codigo(textos, regex):
  if type(textos) == str:
    return regex.sub('CODE', textos)
  else:
    return [regex.sub('CODE', texto) for texto in textos]
```

```python
print(pregunta_español)
```

```python
regex_codigo = re.compile(r'<code>(.|(\n))*?</code>')
```

```python
pregunta_sin_code = sustituir_codigo(pregunta_portugues, regex_codigo)
print(pregunta_sin_code)
```

```python
pregunta_sin_code = sustituir_codigo(pregunta_español, regex_codigo)
print(pregunta_sin_code)
```

```python
pregunta_sin_code = sustituir_codigo(pregunta_ingles, regex_codigo)
print(pregunta_sin_code)
```

#Aula 04

```python
regex_puntuacion = re.compile(r'[^\w\s]')
```

```python
print(eliminar(datos_español.sin_code_tag[0], regex_puntuacion))
```

```python
def minuscula(textos):
  if type(textos) == str:
    return textos.lower()
  else:
    return [texto.lower() for texto in textos]
```

```python
print(minuscula(datos_español.sin_code_tag[0]))
```

```python
#eliminando numeros
regex_digitos = re.compile(r'\d+')
```

```python
print(eliminar('Alura 1234 Latam', regex_digitos))
```

```python
regex_salto_linea = re.compile(r'(\n)')  #saltos de linea
regex_espacio = re.compile(r' +')
```

```python
def  sustituir_por_espacio(textos, regex):
  if type(textos) == str:
    return regex.sub(" ", textos)
  else:
    return [regex.sub(" ", texto) for texto in textos]
```

```python
resultado = (sustituir_por_espacio('Alura 1234  Latam', regex_salto_linea))
print(sustituir_por_espacio(resultado, regex_espacio))
```

```python
preguntas_port_sin_punt = eliminar(datos_portugues.sin_code_tag, regex_puntuacion)
preguntas_port_sin_punt_minus = minuscula(preguntas_port_sin_punt)
preguntas_port_sin_punt_minus_dig = eliminar(preguntas_port_sin_punt_minus, regex_digitos)
preguntas_port_sin_salto_linea = sustituir_por_espacio(preguntas_port_sin_punt_minus, regex_salto_linea)
preguntas_port_sin_espacio_duplicado = sustituir_por_espacio(preguntas_port_sin_salto_linea, regex_espacio)


datos_portugues['preguntas_tratadas'] = preguntas_port_sin_espacio_duplicado
datos_portugues.head()
```

```python
datos_portugues.preguntas_tratadas[0]
```

```python
preguntas_ing_sin_punt = eliminar(datos_ingles.sin_code_tag, regex_puntuacion)
preguntas_ing_sin_punt_minus = minuscula(preguntas_ing_sin_punt)
preguntas_ing_sin_punt_minus_dig = eliminar(preguntas_ing_sin_punt_minus, regex_digitos)
preguntas_ing_sin_salto_linea = sustituir_por_espacio(preguntas_ing_sin_punt_minus, regex_salto_linea)
preguntas_ing_sin_espacio_duplicado = sustituir_por_espacio(preguntas_ing_sin_salto_linea, regex_espacio)


datos_ingles['preguntas_tratadas'] = preguntas_ing_sin_espacio_duplicado
datos_ingles.head()
```
