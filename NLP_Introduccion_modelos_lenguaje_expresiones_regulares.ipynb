#Regex

```python
import re
```

```python
re.findall(r'<.*?>', pregunta_portugues)
```

```python
print(pregunta_portugues)
```

```python
texto_test = re.sub(r'<.*?>','T----E----S----T  ', pregunta_portugues) #removiendo etiquetas
print(texto_test)
```

```python
re.search(r'70','1872873612132629438670')
```

```python
regex = re.compile(r'70')
regex.search('1872873612132629438670')
```

```python
from timeit import timeit   #tiempos de ejecucion
```

```python
setup = '''import re'''
timeit('''re.search(r'70','1872873612132629438670')''', setup)
```

```python
setup = '''import re
regex = re.compile(r'70')'''
timeit('''regex.search('1872873612132629438670')''', setup)   #metodo mas eficiente
```
